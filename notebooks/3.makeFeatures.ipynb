We used three external softwares to generate our features. They were:

1. iFeature
2. POSSUM
3. Kebabs


# iFeature

iFeature was used to generate 21 feature files. It works with the same conda environment defined for our ML tool. We forked the iFeature directory from the original github repo [link] and wrote three bash scripts to generate feature vectors for our enzymes. The bash scripts are given in the forked version of our iFeature repo: [link]. Please note that for the bash scripts to work, the iFeature folder must be in the same directory as the Thioesterase folder. 

The bash scripts were:

1. generateiFeature.sh - takes three arguments, feature_type, input_fasta_file_path, output_directory_to_store_features. Generates feature for the given type of feature extraction technique and stores it in the mentioned directory.
2. generateiFeatures.sh - generates 21 types of features given an input fasta filepath and an output directory to store the sequence. 
3. getPSSMprofile.sh -  for this to work, blastp must be on path. It has two arguments, input_fasta_file_path and output_directory_to_store_files. It will generate the PSSM profiles for the given enzymes in the fasta sequence which will be used later to generate PSSM profile based features. 


We used the following commands to generate features:

1. bash generateIFeatures.sh ../ThioesteraseEnzymeSpecificity/data/seq/EnzymeFasta.fa ../ThioesteraseEnzymeSpecificity/featEngg/offline/ifeatMethods/data/featfiles/trainfiles/

2. bash generateIFeatures.sh ../ThioesteraseEnzymeSpecificity/data/seq/TestEnzymeFasta.fa ../ThioesteraseEnzymeSpecificity/featEngg/offline/ifeatMethods/data/featfiles/testfiles/

3. bash generateIFeatures.sh ../ThioesteraseEnzymeSpecificity/data/seq/ThymeEnzymeFasta1.fa ../ThioesteraseEnzymeSpecificity/featEngg/offline/ifeatMethods/data/featfiles/thymefiles1/

4. bash generateIFeatures.sh ../ThioesteraseEnzymeSpecificity/data/seq/ThymeEnzymeFasta2.fa ../ThioesteraseEnzymeSpecificity/featEngg/offline/ifeatMethods/data/featfiles/thymefiles2/


We used the following commands to generate PSSM profiles:

1. bash getPSSMprofiles.sh ../ThioesteraseEnzymeSpecificity/data/seq/EnzymeFasta.fa ../ThioesteraseEnzymeSpecificity/featEngg/offline/pssmMethods/data/pssmProfiles/trainfiles/

2. bash getPSSMprofiles.sh ../ThioesteraseEnzymeSpecificity/data/seq/TestEnzymeFasta.fa ../ThioesteraseEnzymeSpecificity/featEngg/offline/pssmMethods/data/pssmProfiles/testfiles/

3. bash getPSSMprofiles.sh ../ThioesteraseEnzymeSpecificity/data/seq/ThymeEnzymeFasta1.fa ../ThioesteraseEnzymeSpecificity/featEngg/offline/pssmMethods/data/pssmProfiles/thymefiles1/

4. bash getPSSMprofiles.sh ../ThioesteraseEnzymeSpecificity/data/seq/ThymeEnzymeFasta2.fa ../ThioesteraseEnzymeSpecificity/featEngg/offline/pssmMethods/data/pssmProfiles/thymefiles2/



# POSSUM

The pssm profiles generated using iFeature were used to create 21 pssm profile based feature vectors. POSSUM requires a different environment. The environment file required to create the conda environment is given in the forked version of our POSSUM repository [link]. The original POSSUM repository is present here : [link]. We wrote two bash scripts to generate PSSM based features using POSSUM. The bash scripts are extrememly similar to the iFeature bash scripts. 

The bash scripts were:

1. generatePSSMfeature.sh - takes four arguments, feature_type, input_fasta_file_path, output_directory_to_store_features and the_directory_where_pssm_profiles_are_stored. Generates feature for the given type of feature extraction technique and stores it in the mentioned output directory.
2. generatePSSMfeatures.sh - generates 21 types of features given an input_fasta_filepath, an_output_directory_to_store_the_features and an_input_pssm_dir_where_pssm_profiles are stored. 


We used the following commands to generate PSSM features:

1. bash generatePSSMfeatures.sh ../ThioesteraseEnzymeSpecificity/data/seq/EnzymeFasta.fa ../ThioesteraseEnzymeSpecificity/featEngg/offline/pssmMethods/data/featfiles/trainfiles/ ../ThioesteraseEnzymeSpecificity/featEngg/offline/pssmMethods/data/pssmProfiles/trainfiles/

2. bash generatePSSMfeatures.sh ../ThioesteraseEnzymeSpecificity/data/seq/TestEnzymeFasta.fa ../ThioesteraseEnzymeSpecificity/featEngg/offline/pssmMethods/data/featfiles/testfiles/ ../ThioesteraseEnzymeSpecificity/featEngg/offline/pssmMethods/data/pssmProfiles/testfiles/

3. bash generatePSSMfeatures.sh ../ThioesteraseEnzymeSpecificity/data/seq/ThymeEnzymeFasta1.fa ../ThioesteraseEnzymeSpecificity/featEngg/offline/pssmMethods/data/featfiles/thymefiles1/ ../ThioesteraseEnzymeSpecificity/featEngg/offline/pssmMethods/data/pssmProfiles/thymefiles1/

4. bash generatePSSMfeatures.sh ../ThioesteraseEnzymeSpecificity/data/seq/ThymeEnzymeFasta2.fa ../ThioesteraseEnzymeSpecificity/featEngg/offline/pssmMethods/data/featfiles/thymefiles2/ ../ThioesteraseEnzymeSpecificity/featEngg/offline/pssmMethods/data/pssmProfiles/thymefiles2/



# Kebabs

3 kernel based features were generated using the KeBABs package. It requires a different environment. The environment file required to create the conda environment is given in our github repository. The R script required to generate the features is present in the github repo. We had to build a combined fasta file composed of training and test sequences since Kebabs only returns the kmers that it had encountered on the feature building set of sequences. If feature extraction process was done separately, kebabs will have uneven training and test set feature vector size. 

We used the following command to generate the features:

1. Rscript Kernel-KeBABs.r
