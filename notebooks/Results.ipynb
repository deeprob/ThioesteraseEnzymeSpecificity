```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
file = '../data/results/EnsembleResults/MultiClassEnsemble.csv'

prec = []
accs = []
with open(file,'r') as f:
    for lines in f:
        vals = lines.strip().split(',')
        prec.append(float(vals[1]))
        accs.append(float(vals[2]))
```

```python
np.mean(prec),np.mean(accs)
```

```python
np.median(prec),np.median(accs)
```

```python
plt.hist(prec)
plt.xlabel('Precision Scores')
plt.ylabel('Counts')
plt.title('Precision Score Distribution of Ensemble Model');
```

```python
plt.hist(accs)
plt.xlabel('Accuracy Scores')
plt.ylabel('Counts')
plt.title('Accuracy Score Distribution of Ensemble Model');
```
