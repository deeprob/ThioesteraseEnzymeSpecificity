```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
file = '../data/results/EnsembleResults/MultiClassEnsemble.csv'

prec = []
accs = []
with open(file,'r') as f:
    for lines in f:
        vals = lines.strip().split(',')
        prec.append(float(vals[0]))
        accs.append(float(vals[2]))
```

```python
np.mean(prec),np.mean(accs)
```

```python
np.median(prec),np.median(accs)
```

```python
figure, axes = plt.subplots(nrows=1, ncols=2, figsize=(12,4))

axes[0].hist(prec, color='teal')
axes[0].set_xlabel('Precision Scores')
axes[0].set_title('Precision Score Distribution of Ensemble Model')

axes[1].hist(accs, color='midnightblue')
axes[1].set_xlabel('Accuracy Scores')
axes[1].set_title('Accuracy Score Distribution of Ensemble Model')

figure.text(0, 0.5, 'Counts', va='center', rotation='vertical')
figure.tight_layout()
plt.savefig('../data/results/EnsembleResults/score_dist.png');
```
